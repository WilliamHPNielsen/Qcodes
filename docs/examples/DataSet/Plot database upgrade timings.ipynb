```python
%matplotlib notebook

import matplotlib.pyplot as plt
```

```python
# Results from original version; git hash 4f3232c9d497ba2c270f1bc07d27fb0d51f7005b 

no_of_runs = [10, 50, 250, 1250, 6250, 10000]
sql_query_times = [0.008, 0.035, 0.220, 1.806, 43.986, 135.483]
sql_update_times = [0.001, 0.003, 0.011, 0.044, 0.234, 0.329]
yaml_times = [0.087, 0.403, 1.821, 7.183, 34.319, 56.895]
```

```python
fig, ax = plt.subplots()
ax.plot(no_of_runs, sql_query_times, 'o--', label='SQL queries')
ax.plot(no_of_runs, yaml_times, 'o--', label='Object serialization')
ax.plot(no_of_runs, sql_update_times, 'o--', label='SQL Updates')
#ax.set_xscale('log')
#ax.set_yscale('log')
ax.set_xlabel('No of runs')
ax.set_ylabel('Time consumption (s)')
plt.legend()
```
